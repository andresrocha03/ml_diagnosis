```python
# remover colunas consideradas irrelevantes
data23_valid = data23.drop(drop_columns, axis=1)
data24_valid = data24.drop(drop_columns, axis=1)

# concatenar dados de 23 e 24
data_concat = pd.concat([data23_valid, data24_valid])
```

```python
# Removendo valores nulos das colunas "EVOLUCAO" e "ATE_UF"
data = data_concat.dropna(subset=["ATE_UF"])
data = data.dropna(subset=["EVOLUCAO"])
```

```python
# 3 => outras causas && 9=> ignorado, remover instâncias com esses valores
data = data.drop(data[data['EVOLUCAO'].isin([3, 9])].index)
```

```python
# Podemos ver que os exames de laboratório (LAB_ELIS_1, LAB_ELIS_2, LAB_MICR_1, LAB_MICR_2), aparentemente, deixaram de ser realizados muitas vezes,
# resultando em um número exagerado de instâncias vazias. Dessa forma, preencher com resultados artificais (e.g., média dos valores não nulos) é inviável.
# Único caso válido de se considerar é o LAB_ELIS_1 que, por algum motivo, tem apenas 5,43% das suas instâncias vazias.
for col in data:
  print(f"Column name = {col}     | Amount of null instances = {data[col].isna().sum()}       | Percentage = {data[col].isna().sum() / len(data) * 100}")

data_no_lab = data.drop(["LAB_ELIS_1", "LAB_ELIS_2", "LAB_MICR_1", "LAB_MICR_2"], axis=1)
```

```python
# - Largando a coluna "TP_NOT" porque, embora teoricamente útil, possuímos apenas dados com valor 2 (Notificação individual), tornando a coluna inútil
# - Coluna "DT_NOTIFIC" não faz muito sentido, já que já sabemos que a pessoa está contaminada e a data não nos diz muito
# - Coluna "ATE_DT_INT" é reduntante, já que já temos coluna declarando se a pessoa foi hospitalizada ou não e sua data não acrescenta muita coisa
# - Coluna "SG_UF_NOT" também não faz sentido já que já temos a unidade federativa por outra coluna e, inclusive, não faria sentido já que separamos em dois grupos regionais
# - Coluna de ID ("Unnamed: 0") não nos diz nada
data_final_columns = data_no_lab.drop(["TP_NOT", "DT_NOTIFIC", "ATE_DT_INT", "SG_UF_NOT", "Unnamed: 0"], axis=1)
```

```python
# Apenas 2 instâncias de mulheres grávidas que foram a óbito. Portanto, não faz muito sentido manter a coluna relativa à gravidez
# print(data_no_lab[(data_no_lab["CS_GESTANT"] == 1) & (data_no_lab["EVOLUCAO"] == 2)])
# print(data_no_lab[(data_no_lab["CS_GESTANT"] == 2) & (data_no_lab["EVOLUCAO"] == 2)])
# print(data_no_lab[(data_no_lab["CS_GESTANT"] == 3) & (data_no_lab["EVOLUCAO"] == 2)])
# print(data_no_lab[(data_no_lab["CS_GESTANT"] == 4) & (data_no_lab["EVOLUCAO"] == 2)])
data_final_columns.drop("CS_GESTANT", axis=1, inplace=True)
```

```python
idades_em_anos = ((data_final_columns['NU_IDADE_N'] >= 4000) & (data_final_columns['NU_IDADE_N'] < 5000)).sum()
idades_em_meses = ((data_final_columns['NU_IDADE_N'] >= 3000) & (data_final_columns['NU_IDADE_N'] < 4000)).sum()
idades_em_dias = ((data_final_columns['NU_IDADE_N'] >= 2000) & (data_final_columns['NU_IDADE_N'] < 3000)).sum()
idades_em_horas = ((data_final_columns['NU_IDADE_N'] >= 1000) & (data_final_columns['NU_IDADE_N'] < 2000)).sum()
print(f"Idades em anos: {idades_em_anos}")
print(f"Idades em meses: {idades_em_meses}")
print(f"Idades em dias: {idades_em_dias}")
print(f"Idades em horas: {idades_em_horas}")
```

```python
def calcula_anos_idade(codigo_idade):
  if codigo_idade >= 4000:
    idade_em_anos = codigo_idade - 4000
  elif codigo_idade >= 3000:
    idade_em_anos = float(codigo_idade - 3000)/12
  elif codigo_idade >= 2000:
    idade_em_anos = float(codigo_idade - 2000)/365
  return idade_em_anos
```

```python
data_age_in_years = data_final_columns.copy()
data_age_in_years['NU_IDADE_N'] = data_final_columns['NU_IDADE_N'].apply(calcula_anos_idade)
```

```python
data_age_in_years['NU_IDADE_N'].value_counts()
```

```python
# Criar uma coluna para cada região - SUL, SUD, NOR, NORD, CO

estado_para_regiao = {
    'AC': 'Norte', 'AP': 'Norte', 'AM': 'Norte', 'PA': 'Norte', 'RO': 'Norte', 'RR': 'Norte', 'TO': 'Norte',
    'AL': 'Nordeste', 'BA': 'Nordeste', 'CE': 'Nordeste', 'MA': 'Nordeste', 'PB': 'Nordeste',
    'PE': 'Nordeste', 'PI': 'Nordeste', 'RN': 'Nordeste', 'SE': 'Nordeste',
    'DF': 'Centro_Oeste', 'GO': 'Centro_Oeste', 'MT': 'Centro_Oeste', 'MS': 'Centro_Oeste',
    'ES': 'Sudeste', 'MG': 'Sudeste', 'RJ': 'Sudeste', 'SP': 'Sudeste',
    'PR': 'Sul', 'RS': 'Sul', 'SC': 'Sul'
}

data_age_in_years['regiao'] = data_age_in_years['ATE_UF'].map(estado_para_regiao)
```

```python
# remover coluna que foi encodada
data_age_in_years = data_age_in_years.drop(columns=['ATE_UF'])
data_age_in_years = pd.get_dummies(data_age_in_years, columns=['regiao'])
```

```python
# Converte todas as colunas que começam com 'regiao_' de booleano para inteiro
cols_regiao = [col for col in data_age_in_years.columns if col.startswith('regiao_')]
data_age_in_years[cols_regiao] = data_age_in_years[cols_regiao].astype(int)

data_age_in_years.head()
```

```python
colunas_sintomas = ["CLI_FEBRE", "CLI_MIALGI", "CLI_CEFALE", "CLI_PROST", "CLI_CONGES", "CLI_VOMITO", "CLI_DIARRE", "CLI_ICTERI", "CLI_RENAL", "CLI_RESPIR", "CLI_CARDIA", "CLI_MENING"]
amount_null_sintomas = [57, 102, 152, 190, 237, 145, 203, 164, 223, 214, 255, 267]
i = 0
for sintoma in colunas_sintomas:
  print(f"{sintoma} => {len(data_age_in_years[data_age_in_years[sintoma] == 9.0]) + amount_null_sintomas[i]}")
  i += 1
```

```python
# Dado que soma de valores nulos e ignorados para colunas de sintomas representam no pior caso 5% dos dados, vamos substituir pela moda associada
for sintoma in colunas_sintomas:
  moda = data_age_in_years[sintoma].mode()[0]
  data_age_in_years[sintoma].fillna(moda, inplace=True)
  data_age_in_years[sintoma].replace(9.0, moda, inplace=True)
```

## CONSTRUINDO SPLIT DE TREINO E TESTE

```python
# Construindo o split do dataset entre conjunto de treinamento e conjunto de testes
data_train = data_age_in_years[data_age_in_years["regiao_Sudeste"] == 0]
data_test = data_age_in_years[data_age_in_years["regiao_Sudeste"] == 1]
```

```python
# --- Dados de treinamento ---
# Temos, a priori, 4902 instâncias de sobreviventes e 264 de óbito, de tal forma que esta categoria representa apenas 5% dos dados.
# Optamos, portanto, por remover instâncias aleatórias até que o conjunto tomasse uma proporção de 75% para 25%, ou seja, até que tivéssemos
# apenas 792 instâncias de sobreviventes, resultando em um total de 1056 instâncias.
data_train = data_train.drop(data_train.query('EVOLUCAO == 1').sample(frac=0.83, random_state=42).index)

# --- Dados de teste ---
# Mesma situação, 3868 instâncias de sobreviventes e apenas 172 de óbito. Para manter proporção de 75% para 25%, precisamos remover
# 3180 instâncias de sobreviventes para que tenhamos, no fim, apenas 516.
data_test = data_test.drop(data_test.query("EVOLUCAO == 1").sample(frac=0.86, random_state=42).index)
```

```python
#converter instancias de óbito para 1 e instâncias de cura para 0
data_train['EVOLUCAO'] = data_train['EVOLUCAO'].map({0: 1, 1: 0})
data_test['EVOLUCAO'] = data_test['EVOLUCAO'].map({0: 1, 1: 0})
```

```python
#queremos encontar a distribuição dos dados em relação à variável alvo

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches


#salvar colunas
columns = data_train.columns
#dropar idade (não é binário), evolução( é a própria variável alvo), regiao_Sudeste (coluna possui todos os valores igual a zero no treino)
columns = columns.drop(["NU_IDADE_N", "EVOLUCAO", "regiao_Sudeste"])

#criar um objeto subplot
fig, ax = plt.subplots(figsize=(20,10), nrows = 3, ncols=6, sharey=True)

#para cada coluna selecionada
for idx_plot, col in enumerate(columns):

  #quantificar pessoas da classe 1 que morreram e pessoas da classe 0 que morreram
  death_1 = data_train.query(f"{col}==1 and EVOLUCAO==1")['EVOLUCAO'].shape[0]
  death_0 = data_train.query(f"{col}==0 and EVOLUCAO==1")['EVOLUCAO'].shape[0]

  print(col)
  #salvar valores como frações da quantidade total de instâncias da coluna
  if (data_train.query(f"{col}==1").shape[0]) == 0:
    death_1 = 0
  else:
    death_1 = death_1/(data_train.query(f"{col}==1").shape[0])

  row_1 = {"feature": col,
         "value": 1,
         "deaths": death_1
          }

  if (data_train.query(f"{col}==0").shape[0]) == 0:
    death_0 = 0
  else:
    death_0 = death_0/(data_train.query(f"{col}==0").shape[0])

  #salvar valores como frações da quantidade total de instâncias da coluna
  row_0 = {"feature": col,
         "value": 0,
         "deaths": death_0
          }

  bars = pd.DataFrame([row_1, row_0])

  row = idx_plot//6
  column = idx_plot%6

  #plotar barras de comparação
  sns.barplot(x="value", y="deaths", data=bars, palette=['red', 'blue'], hue='value', ec='k', ax=ax[row][column])



  ax[row][column].set_title(bars["feature"][0])
  ax[row][column].set_xlabel('')
  ax[row][column].set_ylabel('Óbitos (#)')
  ax[row][column].set_xticks([0, 1], (' ', ' '))

#criar objetos para cada modelo
zeros = mpatches.Patch(color='red', label='0')
ones = mpatches.Patch(color='blue', label='1')

fig.legend(handles=[zeros, ones], loc='lower center', bbox_to_anchor=(0.5, 0.05), ncol=4, frameon=False)


```
